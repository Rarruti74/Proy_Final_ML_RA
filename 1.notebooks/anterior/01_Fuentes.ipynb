### Tabla de State y Codigos ZIP de US: "zip_code_database.csv"

El código ZIP (de la sigla en inglés "Zoning Improvement Plan") es un código postal que sirve para identificar áreas geográficas con fines de entrega de correo. Un código FIPS (de la sigla en inglés "Federal Information Processing Standard") es un código numérico utilizado por el gobierno de EE. UU. para identificar de manera única estados, condados y otras subdivisiones geográficas. 

```python
df_ZIPcode = pd.read_csv ("zip_code_database.csv", sep=",")
df_ZIPcode.head()
```

```python
df_zip_code = pd.read_csv ("zip_code_database.csv", sep=",")
df_zip_code = df_zip_code.rename(columns={'state': 'State_ref' ,'primary_city': 'City_ref', 'zip': 'ZIP_ref'})
df_zip = df_zip_code[['ZIP_ref','City_ref' ,'State_ref']]
# print (df_ZIP[df_ZIP['ZIP_ref'] == 43068])

# Guardamos el archivo df_zip.csv
df_zip.to_csv('df_zip.csv', index=False, header=True)
df_zip
```

### Date received and Date sent to company = Delivery time

```python
df['Date received'] = pd.to_datetime(df['Date received'])
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])
df['Delivery Time'] = (df['Date sent to company'] - df['Date received'])
df['Delivery Time'] = df['Delivery Time'].dt.days
print(df.dtypes)
print("-"*40)
print (df['Delivery Time'].sort_values(ascending=True).unique())
print("-"*40)
print (df['Delivery Time'].value_counts(ascending=False))
df = df.drop(columns=['Date received', 'Date sent to company'])
df
```
