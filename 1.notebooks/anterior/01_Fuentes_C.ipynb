### Función: def sustituir_zips_proximos(df_temp, df_ref) -> df_temp

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches

def sustituir_zips_proximos(df_temp, df_ref):
    """
    Sustituye los 'ZIP code' de un df_temp por los más próximos de un df_ref.

    Args:
        df_temp (pd.DataFrame): DataFrame con la columna 'ZIP code' a corregir.
        df_ref (pd.DataFrame): DataFrame de referencia con una columna 'ZIP_ref'
                               que contiene los códigos ZIP correctos.

    Returns:
        pd.DataFrame: El DataFrame df_temp con los 'ZIP code' corregidos.
    """
    # 1. Crear una lista de ZIP codes de referencia
    zip_ref_list = df_ref['ZIP_ref'].dropna().unique().tolist()

    # 2. Convertir la lista de referencia a strings para get_close_matches
    zip_ref_str = [str(int(zip_code)) for zip_code in zip_ref_list]

    # 3. Iterar sobre las filas de df_temp para corregir
    for index, row in df_temp.iterrows():
        current_zip = row['ZIP code']

        # Omitir si el ZIP code actual es NaN
        if pd.isna(current_zip):
            continue

        str_current_zip = str(int(current_zip))

        # Verificar si el ZIP code está en la lista de referencia
        if str_current_zip in zip_ref_str:
            continue  # El ZIP es correcto, no se necesita corrección

        # Buscar la coincidencia más cercana
        # cutoff=0.6 significa que la coincidencia debe tener al menos un 60% de similitud
        closest_matches = get_close_matches(str_current_zip, zip_ref_str, n=1, cutoff=0.6)

        if closest_matches:
            # Si se encuentra una coincidencia, sustituir el valor en el DataFrame
            corrected_zip = int(closest_matches[0])
            df_temp.at[index, 'ZIP code'] = corrected_zip

    return df_temp
```

### Función: def obtener_moda_por_grupo(df, col_moda, col_grupo1, col_grupo2) -> moda_por_grupo

```python
import pandas as pd
import numpy as np

# Función para obtener la moda de una columna agrupando por otras dos
def obtener_moda_por_grupo(df, col_moda, col_grupo1, col_grupo2):
    """
    Calcula la moda de una columna (col_moda) agrupando por dos columnas (col_grupo1, col_grupo2).
    Los valores NaN en las columnas de agrupación son tratados como un grupo único.

    Args:
        df (pd.DataFrame): El DataFrame de entrada.
        col_moda (str): El nombre de la columna para la cual se calculará la moda.
        col_grupo1 (str): El nombre de la primera columna para agrupar.
        col_grupo2 (str): El nombre de la segunda columna para agrupar.

    Returns:
        pd.Series: Una Serie de pandas que contiene la moda de col_moda
                   para cada combinación de col_grupo1 y col_grupo2.
    """
    moda_por_grupo = df.groupby([col_grupo1, col_grupo2], dropna=False)[col_moda].transform(
        lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan).to_dict(orient="index")
    
    return moda_por_grupo
```

### Convertir el df con indice multiple a diccionario para mapeado.

```python
import pandas as pd
import numpy as np

def df_a_dict(df):
    """
    Convierte un DataFrame con índice múltiple (Company, State) y columna 'ZIP code'
    a un diccionario con el formato {Company: {'State': ..., 'ZIP code': ...}}.

    Args:
        df (pd.DataFrame): El DataFrame de entrada con índice de múltiples niveles.

    Returns:
        dict: El diccionario convertido con el formato deseado.
    """
    # 1. Restablecer el índice para que 'Company' y 'State' sean columnas
    df_reset = df.reset_index()

    # 2. Crear un diccionario anidado para cada Company
    # La clave es la 'Company', y el valor es un diccionario con 'State' y 'ZIP code'.
    df_reset.set_index('Company', inplace=True)
    result_dict = {}
    for company, group in df_reset.groupby(level=0):
        # Tomamos el primer valor de 'State' y 'ZIP code' para cada grupo
        # asumiendo que solo hay una entrada por compañía
        state = group['State'].iloc[0]
        zip_code = group['ZIP code'].iloc[0]
        result_dict[company] = {'State': state, 'ZIP code': zip_code}

    return result_dict
```

### Función: def sustituir_nan_con_moda(df_temp, moda_zip_por_grupo, moda_state_por_grupo) -> df_result

```python

def sustituir_nan_con_moda_corregido(df_temp):
    """
    Sustituye los valores NaN de forma eficiente en las columnas 'ZIP code' y 'State'.
    """
    df_result = df_temp.copy()

    # Obtener las modas directamente para la columna 'ZIP code'
    moda_zip = obtener_moda_por_grupo(df_result, 'ZIP code', 'State', 'Company')
    
    # Rellenar NaN en 'ZIP code' con las modas calculadas
    df_result['ZIP code'] = df_result['ZIP code'].fillna(moda_zip)

    # Obtener las modas directamente para la columna 'State'
    # Importante: El cálculo de la moda del State debe hacerse después de
    # corregir el 'ZIP code', para usar los datos más completos.
    moda_state = obtener_moda_por_grupo(df_result, 'State', 'ZIP code', 'Company')

    # Rellenar NaN en 'State' con las modas calculadas
    df_result['State'] = df_result['State'].fillna(moda_state)

    return df_result
```

```python
# Función para sustituir los datos faltantes con la moda de una columna, agrupando otras dos.
def sustituir_nan_con_moda(df_temp, moda_zip_por_grupo, moda_state_por_grupo):
    """
    Sustituye los valores NaN en 'ZIP code' y 'State' de df_temp
    usando las modas precalculadas.

    Args:
        df_temp (pd.DataFrame): DataFrame con valores NaN a sustituir.
        moda_zip_por_grupo (pd.Series): Serie con las modas de 'ZIP code'.
        moda_state_por_grupo (pd.Series): Serie con las modas de 'State'.

    Returns:
        pd.DataFrame: El DataFrame df_temp con los valores NaN sustituidos.
    """
    df_result = df_temp.copy()

    # 1. Rellenar 'ZIP code'
    # Primero, creamos una nueva columna que combina las columnas de grupo
    df_result['group_key'] = list(zip(df_result['State'], df_result['Company']))
    
    # Creamos un diccionario de mapeo a partir de la Serie de modas
    zip_moda_dict = moda_zip_por_grupo.to_dict()

    # Usamos .loc para rellenar solo las filas donde 'ZIP code' es NaN
    df_result.loc[df_result['ZIP code'].isna(), 'ZIP code'] = df_result['group_key'].map(zip_moda_dict)
    
    # 2. Rellenar 'State'
    # Creamos una nueva clave de grupo para el mapeo del Estado
    df_result['group_key_state'] = list(zip(df_result['ZIP code'], df_result['Company']))
    
    # Creamos un diccionario de mapeo a partir de la Serie de modas de 'State'
    state_moda_dict = moda_state_por_grupo.to_dict()

    # Usamos .loc para rellenar solo las filas donde 'State' es NaN
    df_result.loc[df_result['State'].isna(), 'State'] = df_result['group_key_state'].map(state_moda_dict)

    # Limpiamos las columnas temporales
    df_result = df_result.drop(columns=['group_key', 'group_key_state'])

    return df_result
```

### Convertir un dataframe a diccionario

```python
# Función para convertir un DataFrame en un diccionario.
def df_a_diccionario(df, columna_indice, columnas_valor):
    """
    Convierte un DataFrame en un diccionario.

    Args:
        df (pd.DataFrame): El DataFrame de entrada.
        columna_indice (str): Nombre de la columna a usar como clave del diccionario.
        columnas_valor (list): Lista de nombres de columnas para usar como valores del diccionario.

    Returns:
        dict: Un diccionario donde la clave es el valor de la columna_indice
              y el valor es un diccionario con las columnas_valor y sus respectivos valores.
    """
    # 1. Crear un DataFrame temporal con las columnas necesarias
    df_temp = df[[columna_indice] + columnas_valor].copy()
    
    # 2. Eliminar filas con valores faltantes para un mapeo limpio
    df_temp.dropna(inplace=True)
    
    # 3. Eliminar duplicados para asegurar claves únicas
    df_temp.drop_duplicates(subset=[columna_indice], inplace=True)
    
    # 4. Establecer la columna de índice como índice del DataFrame
    df_temp.set_index(columna_indice, inplace=True)
    
    # 5. Convertir a diccionario
    # Se usa orient='index' para obtener un diccionario con el índice como clave
    # y el resto de las columnas como un diccionario anidado.
    diccionario_final = df_temp.to_dict(orient='index')
    
    # 6. Convertir los valores anidados en tuplas para coincidir con el formato deseado
    diccionario_tupla = {clave: tuple(valor.values()) for clave, valor in diccionario_final.items()}
    
    # Aquí es donde puedes elegir el formato.
    # Si quieres un diccionario anidado como el del ejemplo, usa diccionario_final
    # Si quieres el formato de tuplas como el que has indicado, usa diccionario_tupla
    return diccionario_final # Devolvemos el formato de diccionario anidado para mayor claridad y utilidad


# --- Ejemplo de uso ---
# Crear un DataFrame de ejemplo
df_temp4 = df.copy()
df_temp4['ZIP code'] = df_temp4['ZIP code'].astype('Int64') # Int64 soporta NaN

print("DataFrame Original:")
print (df_temp4)

# Definir las columnas para el mapeo
columna_clave = 'Company'
columnas_valor = ['State', 'ZIP code']

# Llamar a la función
mapeo_completo2 = df_a_diccionario(df_temp4, columna_clave, columnas_valor)

print("\nDiccionario resultante:")
mapeo_completo2

# Verificación de dtype (ejemplo: accediendo a un valor)
print(f"\nTipo de dato para 'ZIP code' de 'Navy FCU': {type(mapeo_completo2['Navy FCU']['ZIP code'])}")
```

### Funcion completa

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches

def zip_corrector(df_main, df_zip, dict_mapeo):
    """
    Corrige y completa los datos de 'ZIP code' y 'State' en df_main.

    Args:
        df_main (pd.DataFrame): DataFrame principal con datos faltantes y erróneos.
        df_zip (pd.DataFrame): DataFrame de referencia con 'State_ref' y 'ZIP_ref'.
        dict_mapeo (dict): Diccionario de referencia con 'Company' como clave
                           y un diccionario de 'State' y 'ZIP code' como valor.

    Returns:
        pd.DataFrame: El DataFrame principal con los datos corregidos.
    """
    # 1. Preparar un DataFrame de referencia completo
    df_ref_from_zip = df_zip.rename(columns={'State_ref': 'State', 'ZIP_ref': 'ZIP code'})
    
    # Crear un DataFrame a partir de dict_mapeo para poder concatenarlo
    df_ref_from_map = pd.DataFrame.from_dict(dict_mapeo, orient='index').reset_index().rename(columns={'index': 'Company'})

    # Combinar todas las referencias
    df_reference = pd.concat([df_ref_from_zip, df_ref_from_map], ignore_index=True).drop_duplicates(subset=['Company', 'State', 'ZIP code'])

    # 2. Iterar y aplicar la lógica de corrección
    for index, row in df_main.iterrows():
        # Caso 1: 'State' es NaN, 'ZIP code' no coincide
        if pd.isna(row['State']) and row['ZIP code'] not in df_reference['ZIP code'].values:
            zip_candidatos = df_reference[df_reference['Company'] == row['Company']]['ZIP code'].tolist()
            if zip_candidatos:
                # Encuentra la coincidencia más cercana
                candidato_cercano = get_close_matches(str(row['ZIP code']), [str(zc) for zc in zip_candidatos], n=1)
                if candidato_cercano:
                    df_main.at[index, 'ZIP code'] = candidato_cercano[0]
                    # Rellenar State después de corregir el ZIP
                    df_main.at[index, 'State'] = df_reference[(df_reference['Company'] == row['Company']) & (df_reference['ZIP code'] == candidato_cercano[0])]['State'].iloc[0]

        # Caso 2: 'State' no es NaN, 'ZIP code' no coincide
        elif not pd.isna(row['State']) and row['ZIP code'] not in df_reference['ZIP code'].values:
            zip_candidatos = df_reference[df_reference['State'] == row['State']]['ZIP code'].tolist()
            if zip_candidatos:
                candidato_cercano = get_close_matches(str(row['ZIP code']), [str(zc) for zc in zip_candidatos], n=1)
                if candidato_cercano:
                    df_main.at[index, 'ZIP code'] = candidato_cercano[0]

        # Caso 3: 'ZIP code' es NaN, 'State' coincide
        elif pd.isna(row['ZIP code']) and row['State'] in df_reference['State'].values:
            # Encontrar la moda de ZIP codes para la Company y el State
            moda_zip = df_reference[(df_reference['Company'] == row['Company']) & (df_reference['State'] == row['State'])]['ZIP code'].mode()
            if not moda_zip.empty:
                df_main.at[index, 'ZIP code'] = moda_zip.iloc[0]

        # Caso 4: 'State' es NaN, 'ZIP code' coincide
        elif pd.isna(row['State']) and row['ZIP code'] in df_reference['ZIP code'].values:
            # Encontrar la moda de States para la Company y el ZIP code
            moda_state = df_reference[(df_reference['Company'] == row['Company']) & (df_reference['ZIP code'] == row['ZIP code'])]['State'].mode()
            if not moda_state.empty:
                df_main.at[index, 'State'] = moda_state.iloc[0]
        
        # Caso 5: 'ZIP code' y 'State' son NaN
        elif pd.isna(row['ZIP code']) and pd.isna(row['State']):
            if row['Company'] in dict_mapeo:
                df_main.at[index, 'ZIP code'] = dict_mapeo[row['Company']].get('ZIP code')
                df_main.at[index, 'State'] = dict_mapeo[row['Company']].get('State')
    
    return df_main
```

### Funcion de cada caso independiente 

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches

def _caso_1(df_main_row, df_ref_zip, df_ref_map):
    """
    Caso 1: 'State' es NaN, 'ZIP code' no coincide.
    Modifica el ZIP code erróneo con el más parecido para la misma 'Company'.
    """
    company_zips = df_ref_map[df_ref_map['Company'] == df_main_row['Company']]['ZIP code'].tolist()
    if company_zips:
        closest_zip = get_close_matches(str(df_main_row['ZIP code']), [str(zc) for zc in company_zips], n=1)
        if closest_zip:
            corrected_zip = int(closest_zip[0])
            corrected_state = df_ref_map[(df_ref_map['Company'] == df_main_row['Company']) & (df_ref_map['ZIP code'] == corrected_zip)]['State'].iloc[0]
            return corrected_state, corrected_zip
    return df_main_row['State'], df_main_row['ZIP code']

def _caso_2(df_main_row, df_ref_zip, df_ref_map):
    """
    Caso 2: 'State' no es NaN, 'ZIP code' no coincide.
    Modifica el ZIP code erróneo con el más parecido para el mismo 'State'.
    """
    state_zips = df_ref_zip[df_ref_zip['State_ref'] == df_main_row['State']]['ZIP_ref'].tolist()
    if state_zips:
        closest_zip = get_close_matches(str(df_main_row['ZIP code']), [str(zc) for zc in state_zips], n=1)
        if closest_zip:
            return df_main_row['State'], int(closest_zip[0])
    return df_main_row['State'], df_main_row['ZIP code']

def _caso_3(df_main_row, df_ref_zip, df_ref_map):
    """
    Caso 3: 'ZIP code' es NaN, 'State' coincide.
    Completa el NaN con la moda del 'ZIP code' para los pares (Company, State).
    """
    company_state_data = pd.concat([df_ref_zip.rename(columns={'State_ref': 'State', 'ZIP_ref': 'ZIP code'}),
                                    df_ref_map], ignore_index=True)
    
    zip_mode = company_state_data[(company_state_data['Company'] == df_main_row['Company']) &
                                  (company_state_data['State'] == df_main_row['State'])]['ZIP code'].mode()
    if not zip_mode.empty:
        return df_main_row['State'], zip_mode.iloc[0]
    return df_main_row['State'], df_main_row['ZIP code']

def _caso_4(df_main_row, df_ref_zip, df_ref_map):
    """
    Caso 4: 'State' es NaN, 'ZIP code' coincide.
    Completa el NaN con la moda del 'State' para los pares (Company, ZIP code).
    """
    company_zip_data = pd.concat([df_ref_zip.rename(columns={'State_ref': 'State', 'ZIP_ref': 'ZIP code'}),
                                  df_ref_map], ignore_index=True)
    
    state_mode = company_zip_data[(company_zip_data['Company'] == df_main_row['Company']) &
                                  (company_zip_data['ZIP code'] == df_main_row['ZIP code'])]['State'].mode()
    if not state_mode.empty:
        return state_mode.iloc[0], df_main_row['ZIP code']
    return df_main_row['State'], df_main_row['ZIP code']
    
def _caso_5(df_main_row, df_ref_zip, df_ref_map):
    """
    Caso 5: 'ZIP code' y 'State' son NaN.
    Completa los NaN con los valores del 'dict_mapeo'.
    """
    if df_main_row['Company'] in df_ref_map['Company'].values:
        mapping_data = df_ref_map[df_ref_map['Company'] == df_main_row['Company']].iloc[0]
        return mapping_data['State'], mapping_data['ZIP code']
    return df_main_row['State'], df_main_row['ZIP code']
```

```python
def zip_corrector(df_main, df_zip, dict_mapeo):
    """
    Orquesta la corrección de datos en df_main aplicando la lógica de cada caso.
    """
    # Preparar DataFrames de referencia para evitar conversiones repetidas
    df_ref_zip = df_zip.copy()
    df_ref_map = pd.DataFrame.from_dict(dict_mapeo, orient='index').reset_index().rename(columns={'index': 'Company'})
    
    df_temp = df_main.copy()
    
    # Aplicar la lógica de cada caso en el orden especificado
    # Caso 1
    mask_1 = df_temp['State'].isna() & ~df_temp['ZIP code'].isin(df_ref_zip['ZIP_ref'])
    df_temp.loc[mask_1, ['State', 'ZIP code']] = df_temp[mask_1].apply(
        lambda row: _caso_1(row, df_ref_zip, df_ref_map), axis=1, result_type='expand'
    )
    
    # Caso 2
    mask_2 = ~df_temp['State'].isna() & ~df_temp['ZIP code'].isin(df_ref_zip['ZIP_ref'])
    df_temp.loc[mask_2, ['State', 'ZIP code']] = df_temp[mask_2].apply(
        lambda row: _caso_2(row, df_ref_zip, df_ref_map), axis=1, result_type='expand'
    )
    
    # Caso 3
    mask_3 = df_temp['ZIP code'].isna() & df_temp['State'].isin(df_ref_zip['State_ref'])
    df_temp.loc[mask_3, ['State', 'ZIP code']] = df_temp[mask_3].apply(
        lambda row: _caso_3(row, df_ref_zip, df_ref_map), axis=1, result_type='expand'
    )
    
    # Caso 4
    mask_4 = df_temp['State'].isna() & ~df_temp['ZIP code'].isna() & df_temp['ZIP code'].isin(df_ref_zip['ZIP_ref'])
    df_temp.loc[mask_4, ['State', 'ZIP code']] = df_temp[mask_4].apply(
        lambda row: _caso_4(row, df_ref_zip, df_ref_map), axis=1, result_type='expand'
    )
    
    # Caso 5
    mask_5 = df_temp['State'].isna() & df_temp['ZIP code'].isna()
    df_temp.loc[mask_5, ['State', 'ZIP code']] = df_temp[mask_5].apply(
        lambda row: _caso_5(row, df_ref_zip, df_ref_map), axis=1, result_type='expand'
    )
    
    return df_temp
```

### CORRECTOR PARA ZIP

```python
import pandas as pd
import numpy as np
from difflib import get_close_matches

def correct_erroneous_zip_codes(df_temp, df_reference):
    """
    Detecta y corrige ZIP codes erróneos en df_temp, reemplazándolos por el más parecido
    de los ZIP codes correctos basados en el State y Company de referencia.

    Args:
        df_temp (pd.DataFrame): DataFrame con la columna 'ZIP code' a corregir.
                                Debe contener también 'State' y 'Company'.
        df_reference (pd.DataFrame): DataFrame de referencia con 'State', 'ZIP code', y 'Company'
                                     sin datos faltantes ni duplicados.

    Returns:
        pd.DataFrame: El DataFrame df_temp con los ZIP codes corregidos.
    """
    df_corrected = df_temp.copy()

    # Asegurarse de que df_reference tiene los tipos de datos correctos para comparación
    df_reference['ZIP code'] = pd.to_numeric(df_reference['ZIP code'], errors='coerce')
    df_corrected['ZIP code'] = pd.to_numeric(df_corrected['ZIP code'], errors='coerce')

    # Convertir ZIP codes de referencia a strings para get_close_matches
    # y filtrar NaN para asegurar que solo se usan ZIPs válidos para la comparación
    valid_ref_zips = df_reference.dropna(subset=['ZIP code'])

    for index, row in df_corrected.iterrows():
        current_zip = row['ZIP code']
        current_state = row['State']
        current_company = row['Company']

        # Solo procesar si el ZIP code actual no es NaN
        if pd.isna(current_zip):
            continue

        # Convertir el ZIP code actual a string para comparación
        str_current_zip = str(int(current_zip)) if not pd.isna(current_zip) else ''

        # Verificar si el ZIP code actual es correcto según la referencia
        # Un ZIP es "correcto" si existe una entrada en df_reference con esa Company, State y ZIP.
        # Priorizamos Company y State para la validación.
        is_correct = False
        if not pd.isna(current_state):
            if ((valid_ref_zips['Company'] == current_company) &
                (valid_ref_zips['State'] == current_state) &
                (valid_ref_zips['ZIP code'] == current_zip)).any():
                is_correct = True
            elif ((valid_ref_zips['State'] == current_state) &
                  (valid_ref_zips['ZIP code'] == current_zip)).any():
                is_correct = True
        elif ((valid_ref_zips['Company'] == current_company) &
              (valid_ref_zips['ZIP code'] == current_zip)).any():
            is_correct = True
        
        if is_correct:
            continue # Si el ZIP code ya es correcto, pasamos a la siguiente fila

        # Si llegamos aquí, el ZIP code es erróneo o no está en la combinación esperada.
        # Buscamos candidatos a ZIP code correcto.
        possible_zips = []

        # Estrategia 1: Buscar por Company y State (si ambos están presentes)
        if not pd.isna(current_company) and not pd.isna(current_state):
            candidates = valid_ref_zips[(valid_ref_zips['Company'] == current_company) &
                                        (valid_ref_zips['State'] == current_state)]['ZIP code'].tolist()
            if candidates:
                possible_zips.extend([str(int(z)) for z in candidates if not pd.isna(z)])

        # Estrategia 2: Si no encontramos por Company y State, buscar solo por Company
        if not possible_zips and not pd.isna(current_company):
            candidates = valid_ref_zips[valid_ref_zips['Company'] == current_company]['ZIP code'].tolist()
            if candidates:
                possible_zips.extend([str(int(z)) for z in candidates if not pd.isna(z)])

        # Estrategia 3: Si aún no encontramos, buscar solo por State (si está presente)
        if not possible_zips and not pd.isna(current_state):
            candidates = valid_ref_zips[valid_ref_zips['State'] == current_state]['ZIP code'].tolist()
            if candidates:
                possible_zips.extend([str(int(z)) for z in candidates if not pd.isna(z)])

        # Si tenemos candidatos, encontrar el más parecido
        if possible_zips:
            # Eliminar duplicados y asegurarse de que son únicos para get_close_matches
            possible_zips_unique = list(set(possible_zips))
            
            # get_close_matches devuelve una lista de coincidencias, la primera es la más cercana
            closest_match = get_close_matches(str_current_zip, possible_zips_unique, n=1, cutoff=0.6) # Ajusta cutoff si necesitas más o menos similitud

            if closest_match:
                corrected_zip = int(closest_match[0])
                df_corrected.at[index, 'ZIP code'] = corrected_zip
                # Opcional: Si el State estaba NaN, y se corrigió el ZIP, podríamos intentar rellenar el State también
                if pd.isna(df_corrected.at[index, 'State']):
                    corresponding_state = valid_ref_zips[valid_ref_zips['ZIP code'] == corrected_zip]['State'].mode()
                    if not corresponding_state.empty:
                        df_corrected.at[index, 'State'] = corresponding_state.iloc[0]
            else:
                # Si no se encontró ninguna coincidencia cercana, podríamos dejarlo como NaN
                # o decidir otra estrategia (ej: loguear el error, mantener el original, etc.)
                df_corrected.at[index, 'ZIP code'] = np.nan # O dejar row['ZIP code'] si prefieres no borrarlo
        else:
            # Si no hay candidatos de referencia para corregir el ZIP, lo dejamos como NaN
            df_corrected.at[index, 'ZIP code'] = np.nan

    return df_corrected

# --- DATOS DE EJEMPLO PARA TESTEAR ---

# DataFrame de referencia (limpio, sin NaNs ni duplicados en 'Company', 'State', 'ZIP code')
# Este sería tu 'dict_mapeo' transformado o una combinación de 'df_zip' y 'dict_mapeo'
df_reference_data = {
    'Company': ['Apple', 'Apple', 'Microsoft', 'Microsoft', 'Google', 'Google', 'Tesla'],
    'State': ['NY', 'CA', 'FL', 'TX', 'WA', 'TX', 'CA'],
    'ZIP code': [10001, 90210, 33150, 75201, 98101, 75001, 90211]
}
df_reference = pd.DataFrame(df_reference_data)

# DataFrame de ejemplo con errores en 'ZIP code'
df_temp_data = {
    'State': ['NY', 'TX', 'CA', 'FL', 'TX', 'NY', 'CA', 'CA'],
    'ZIP code': [10002, 75002, 90210, 33150, 75005, 10003, 90215, 12345], # 10002, 75002, 75005, 10003, 90215, 12345 son erróneos
    'Company': ['Apple', 'Google', 'Apple', 'Microsoft', 'Google', 'Apple', 'Tesla', 'Unknown']
}
df_temp = pd.DataFrame(df_temp_data)

print("DataFrame temp original:")
print(df_temp)
print("\n" + "="*50 + "\n")

print("DataFrame de referencia:")
print(df_reference)
print("\n" + "="*50 + "\n")

# Aplicar la función de corrección
df_temp_corrected = correct_erroneous_zip_codes(df_temp, df_reference)

print("DataFrame temp corregido:")
print(df_temp_corrected)
```
