(API): https://api.zip-codes.com/ZipCodesAPI.svc/1.0/<endpoint>?key=<APIKEY>

```python
import pandas as pd
import requests

def verificar_codigos_postales(df, columna_zip, api_url):
    """
    Verifica códigos postales en un DataFrame usando una API.

    Args:
        df: DataFrame de pandas.
        columna_zip: Nombre de la columna que contiene los códigos postales.
        api_url: URL de la API que valida los códigos postales.

    Returns:
        DataFrame con una nueva columna indicando si el código postal es válido.
    """
    def es_zip_valido(zip_code):
        try:
            response = requests.get(f"{api_url}{zip_code}")
            response.raise_for_status()  # Lanza una excepción para códigos de error HTTP
            data = response.json()
            return data.get("es_valido", False)  # Ajusta según la estructura de la respuesta de tu API
        except requests.exceptions.RequestException:
            return False  # Manejo de errores de la API

    df['zip_valido'] = df[columna_zip].apply(es_zip_valido)
    return df

# Ejemplo de uso
# Asumiendo que tienes un DataFrame llamado 'datos' con una columna 'codigo_postal'
# y una API en 'https://api.example.com/zipcodes/'
# datos = pd.read_csv('archivo.csv') # Cargar datos desde un archivo
# datos_verificados = verificar_codigos_postales(datos.copy(), 'codigo_postal', 'https://api.example.com/zipcodes/')
# print(datos_verificados)
```

### Función: def sustituir_zips_proximos(df_temp, df_ref) -> df_temp

```python
def sustituir_zips_proximos(df_temp, df_ref):
    """
    Sustituye los 'ZIP code' de un df_temp por los más próximos de un df_ref.

    Args:
        df_temp (pd.DataFrame): DataFrame con la columna 'ZIP code' a corregir.
        df_ref (pd.DataFrame): DataFrame de referencia con una columna 'ZIP_ref'
                               que contiene los códigos ZIP correctos.

    Returns:
        pd.DataFrame: El DataFrame df_temp con los 'ZIP code' corregidos.
    """
    # 1. Crear una lista de ZIP codes de referencia
    zip_ref_list = df_ref['ZIP_ref'].dropna().unique().tolist()

    # 2. Convertir la lista de referencia a strings para get_close_matches
    zip_ref_str = [str(int(zip_code)) for zip_code in zip_ref_list]

    # 3. Iterar sobre las filas de df_temp para corregir
    for index, row in df_temp.iterrows():
        current_zip = row['ZIP code']

        # Omitir si el ZIP code actual es NaN
        if pd.isna(current_zip):
            continue

        str_current_zip = str(int(current_zip))

        # Verificar si el ZIP code está en la lista de referencia
        if str_current_zip in zip_ref_str:
            continue  # El ZIP es correcto, no se necesita corrección

        # Buscar la coincidencia más cercana
        # Hemos definido cutoff=0.5 para que la coincidencia sea al menos de un 50% de similitud. Asi obtenemos un valor para 'ZIP code' = 11
        from difflib import get_close_matches
        closest_matches = get_close_matches(str_current_zip, zip_ref_str, n=1, cutoff=0.5)

        if closest_matches:
            # Si se encuentra una coincidencia, sustituir el valor en el DataFrame
            corrected_zip = int(closest_matches[0])
            df_temp.at[index, 'ZIP code'] = corrected_zip

    return df_temp
```

### Función: def obtener_moda_por_grupo(df, col_moda, col_grupo1, col_grupo2) -> moda_por_grupo

```python
# Función para obtener la moda de una columna agrupando por otras dos
def obtener_moda_por_grupo(df, col_moda, col_grupo1, col_grupo2):
    """
    Calcula la moda de una columna (col_moda) agrupando por dos columnas (col_grupo1, col_grupo2).
    Los valores NaN en las columnas de agrupación son tratados como un grupo único.

    Args:
        df (pd.DataFrame): El DataFrame de entrada.
        col_moda (str): El nombre de la columna para la cual se calculará la moda.
        col_grupo1 (str): El nombre de la primera columna para agrupar.
        col_grupo2 (str): El nombre de la segunda columna para agrupar.

    Returns:
        pd.Series: Una Serie de pandas que contiene la moda de col_moda
                   para cada combinación de col_grupo1 y col_grupo2.
    """
    moda_por_grupo = df.groupby([col_grupo1, col_grupo2], dropna=False)[col_moda].transform(
        lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan).to_dict(orient="index")
    
    return moda_por_grupo
```

### Convertir el df con indice multiple a diccionario para mapeado.

```python

def df_a_dict(df):
    """
    Convierte un DataFrame con índice múltiple (Company, State) y columna 'ZIP code'
    a un diccionario con el formato {Company: {'State': ..., 'ZIP code': ...}}.

    Args:
        df (pd.DataFrame): El DataFrame de entrada con índice de múltiples niveles.

    Returns:
        dict: El diccionario convertido con el formato deseado.
    """
    # 1. Restablecer el índice para que 'Company' y 'State' sean columnas
    df_reset = df.reset_index()

    # 2. Crear un diccionario anidado para cada Company
    # La clave es la 'Company', y el valor es un diccionario con 'State' y 'ZIP code'.
    df_reset.set_index('Company', inplace=True)
    dict_zip_moda = {}
    for company, group in df_reset.groupby(level=0):
        # Tomamos el primer valor de 'State' y 'ZIP code' para cada grupo
        # asumiendo que solo hay una entrada por compañía
        state = group['State'].iloc[0]
        zip_code = group['ZIP code'].iloc[0]
        dict_zip_moda[company] = {'State': state, 'ZIP code': zip_code}

    return dict_zip_moda
```

### Función: def sustituir_nan_con_moda(df_temp, moda_zip_por_grupo, moda_state_por_grupo) -> df_result

```python
# Función para sustituir los datos faltantes con la moda de una columna, agrupando otras dos.
def sustituir_nan_con_moda(df_temp, moda_zip_por_grupo, moda_state_por_grupo):
    """
    Sustituye los valores NaN en 'ZIP code' y 'State' de df_temp
    usando las modas precalculadas.

    Args:
        df_temp (pd.DataFrame): DataFrame con valores NaN a sustituir.
        moda_zip_por_grupo (pd.Series): Serie con las modas de 'ZIP code'.
        moda_state_por_grupo (pd.Series): Serie con las modas de 'State'.

    Returns:
        pd.DataFrame: El DataFrame df_temp con los valores NaN sustituidos.
    """
    df_result = df_temp.copy()

    # 1. Rellenar 'ZIP code'
    # Primero, creamos una nueva columna que combina las columnas de grupo
    df_result['group_key'] = list(zip(df_result['State'], df_result['Company']))
    
    # Creamos un diccionario de mapeo a partir de la Serie de modas
    zip_moda_dict = moda_zip_por_grupo.to_dict()

    # Usamos .loc para rellenar solo las filas donde 'ZIP code' es NaN
    df_result.loc[df_result['ZIP code'].isna(), 'ZIP code'] = df_result['group_key'].map(zip_moda_dict)
    
    # 2. Rellenar 'State'
    # Creamos una nueva clave de grupo para el mapeo del Estado
    df_result['group_key_state'] = list(zip(df_result['ZIP code'], df_result['Company']))
    
    # Creamos un diccionario de mapeo a partir de la Serie de modas de 'State'
    state_moda_dict = moda_state_por_grupo.to_dict()

    # Usamos .loc para rellenar solo las filas donde 'State' es NaN
    df_result.loc[df_result['State'].isna(), 'State'] = df_result['group_key_state'].map(state_moda_dict)

    # Limpiamos las columnas temporales
    df_result = df_result.drop(columns=['group_key', 'group_key_state'])

    return df_result
```
